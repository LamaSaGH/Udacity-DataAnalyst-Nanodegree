## Investigation Overview


> The goal of this investigation is to understand user behavior and usage patterns in a bike-sharing system, focusing on the differences between Subscribers and Customers. By analyzing trip duration, user type, start time, and other relevant factors.


## Dataset Overview and Executive Summary

>This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area



```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
Data=pd.read_csv('fordgobike-tripdata.csv')
sns.set_palette('mako')
```

## Distribution of User Type

> Subscribers make significantly more trips compared to Customers.

```python
plt.figure(figsize=(10,8))
sns.countplot(data=Data, x='user_type')

plt.title('Distribution of User Type')
plt.xlabel('User Type')
plt.ylabel('Count')

plt.show()
```

## Bike Share Usage by User Type

> Subscribers tend to use the bike share for all trips more frequently than Customers.


```python
plt.figure(figsize=(10, 6))
sns.countplot(x='user_type', hue='bike_share_for_all_trip', data=Data)
plt.title('Bike Share Usage by User Type')
plt.xlabel('User Type')
plt.ylabel('Count')
plt.legend(title='Bike Share for All Trip')
plt.show()
```

## Trip Duration (minutes) Distribution by User Type

> Subscribers generally have shorter and more consistent trip durations, with most trips under 20 minutes. Customers show a wider range of trip durations, with many longer trips.

```python
Data['duration_min'] = Data['duration_sec'] / 60
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='user_type', y='duration_min', data=Data)
plt.title('Trip Duration (minutes) Distribution by User Type')
plt.xlabel('User Type')
plt.ylabel('Duration (minutes)')
plt.ylim(0, 100) 
plt.show()
```
